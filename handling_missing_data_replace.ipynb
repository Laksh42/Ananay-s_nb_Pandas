### Replacing Single Value

```python
df
```

```python
new_df=df.replace(-99999,value=np.NaN)
new_df
```

```python
new_df.info()
```

### Replacing List with Single Value

```python
new_df=df.replace([-99999,-888888],value=0)
new_df
```

```python
new_df.info()
```

### Replacing Per Column

```python
df
```

```python
new_df=df.replace({'temperature':-99999,'windspeed':-99999,'event':'0'},np.nan)
new_df
```

```python
new_df.info()
```
